# Basic operations

This notebook is an adaptation of the tutorial on https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_core/py_basic_ops/py_basic_ops.html#basic-ops. It covers how to:

* Access and modify specific pixels
* Access Regions of an Image

````python
import numpy as np
import cv2
from matplotlib import pyplot as plt
````

### Accessing and modifying individual pixels

For a grayscale image

````python
# specify the filepath and read the image
fp = 'C:/Users/20194136/Documents/Personal_Development/opencv/Data/bike.jpg'
# the zero argument here makes it grayscale
img = cv2.imread(fp, 0)
````

````python
# Let's access the pixel at a specific row and column
row = 389
column = 544
pixel = img[row,column]
print(pixel)
````

````python
# Let's now change this pixel into a different color
img[row,column] = 180
````

````python
# plot the image with the changed pixel
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
````

We can also edit the pixel in the following way:

````python
img.itemset((row,column), 20)
````

However, we currently don't see much of a change, since only one small pixel is changed. Let's see if we can access different regions.

### Accessing regions of an image.

````python
# Let's print the image shape again
img.shape
````

````python
# To find the part with the bike, plot different parts until you 
# Have selected only the bike. 
plt.imshow(img[245:400,185:460], cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
````

````python
# Let's try to take the region of the image that has the bike, and 
# copy it to a different section of the image.

bike = img[245:400, 185:460]
img_two_bikes = img
img_two_bikes[145:300, 185:460] = bike
````

````python
# Now we have two bikes on top of each other! 
plt.imshow(img_two_bikes, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
````

### Image addition

````python
# Let's load two images now in color
fp1 = 'C:/Users/20194136/Documents/Personal_Development/opencv/Data/bike.jpg'
bike = cv2.imread(fp1)
bike = bike[:, :, [2, 1, 0]]

fp2 = 'C:/Users/20194136/Documents/Personal_Development/opencv/Data/mountain.jpg'
mountain = cv2.imread(fp2)
mountain = mountain[:, :, [2, 1, 0]]
````

````python
bike_mountain = bike+mountain
# or alternatively you can use
#bike_mountain = cv2.add(bike, mountain)
````

````python
# Let's plot
plt.imshow(bike_mountain, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
````

### Image blending

Now, different weights are given to images so that the two images kind of blend into each other. You can blend images using ```cv2.addWeighted```

````python
bike_mountain_blend = cv2.addWeighted(bike,0.5, mountain,0.5, 0)

#plot!
plt.imshow(bike_mountain_blend, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
````
