# Reading and Saving images

This notebook is about how to read and save images.

```python
# import numpy and opencv
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# specify the filepath and read the image
fp = 'C:/Users/20194136/Documents/Personal_Development/opencv/Data/bike.jpg'
# the zero argument here makes it grayscale
img = cv2.imread(fp, 0)
```

### Reading the image

```python
# show the image in grayscale
cv2.imshow('bike', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# plot the image with matplotlib
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

### Reading color image

Now we can read the same image, but take colors instead of grayscale. 

```python
img_color = cv2.imread(fp)
```

```python
# show the image in color
cv2.imshow('bike', img_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

However, when we want to plot this same picture in matplotlib, we see that the colors are really strange! What happened? OpenCV uses BGR (Blue/Green/Red) to identify the colors, whereas matplotlib and Pillow use the more commonly known RGB (Red/Green/Blue). Therefore, to read it properly, we need to convert BGR to RGB. 

```python
# for clarity, let's rename our img_color to img_color_bgr
img_color_bgr = img_color
```

```python
# plot the image with matplotlib
plt.imshow(img_color_bgr, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# What we do is simply change the dimensions from BGR to RGB 
#(so swap the first and last)
img_color_rgb = img_color_bgr[:, :, [2, 1, 0]]
```

```python
# plot the image with matplotlib
plt.imshow(img_color_rgb, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
